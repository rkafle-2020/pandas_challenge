```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete['school_name'].unique())
total_students = school_data_complete['student_name'].count()
total_budget = school_data['budget'].sum()
average_math_score = school_data_complete['math_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()

passing_students_math = school_data_complete[school_data_complete['math_score'] >=70]['student_name'].count()
passing_rate_math = (passing_students_math / total_students) * 100

passing_students_reading = school_data_complete[school_data_complete['reading_score'] >= 70]['student_name'].count()
passing_rate_reading = (passing_students_reading / total_students) * 100

overall_passing_rate = school_data_complete[(school_data_complete['math_score'] >=70) & 
                                                (school_data_complete['reading_score'] >=70)]['student_name'].count()
Passing_rate_overall = overall_passing_rate/total_students *100
```

```python
District_summary = pd.DataFrame({"Total Schools": [total_schools],
                                "Total Students": [total_students],
                                "Total Budget": [total_budget],
                                "Average Math Score": [average_math_score],
                                "Average Reading Score": [average_reading_score],
                                "Passing Rate Math": [passing_rate_math],
                                "Passing Rate Reading": [passing_rate_reading],
                                "Overall Passing Rate": [Passing_rate_overall]})


District_summary.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Calculating the summary metrics as per above

School_Name = school_data["school_name"] 
school_type = school_data_complete['type']
student_per_school = school_data_complete['school_name'].value_counts()
total_school_budget = school_data_complete.groupby('school_name')['budget'].mean()
per_student_budget = total_school_budget/student_per_school
avg_math_score = school_data_complete.groupby('school_name')['math_score'].mean()
avg_reading_score= school_data_complete.groupby('school_name')['reading_score'].mean()

#Calculating percentage passing Math, Reading and Overall
total_stu_passing_math = school_data_complete[school_data_complete['math_score'] >=70].groupby(['school_name'])['student_name'].count()
Percent_passing_math = total_stu_passing_math/student_per_school*100
total_stu_passing_reading = school_data_complete[school_data_complete['reading_score'] >=70].groupby(['school_name'])['student_name'].count()
Percent_passing_reading = total_stu_passing_reading/student_per_school*100
overall_passing_rate = school_data_complete[(school_data_complete['math_score'] >=70) & 
                                                (school_data_complete['reading_score'] >=70)].groupby(['school_name'])['student_name'].count()/student_per_school*100

#Creating the data frame
School_Summary = pd.DataFrame({"School Name": [School_Name],
                              "School Type": [school_type],
                              "Total Students": [student_per_school],
                              "Total School Budget": [total_school_budget],
                              "Per Student Budget": [per_student_budget],
                              "Average Math Score": [avg_math_score],
                              "Average Reading Score": [avg_reading_score],
                              "% Passing Math": [Percent_passing_math],
                               "% Passing Reading": [Percent_passing_reading],
                               "% Overall Passing": [overall_passing_rate]})
School_Summary.set_index('School Name', inplace=True)

School_Summary.head(10)                               
                              


                                                                             
```
